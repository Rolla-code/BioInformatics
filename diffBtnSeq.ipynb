```python
""" Finding the difference between sequences"""

    
with open('C:/Users/white/Desktop/seq1.fasta', 'r') as f1: #put your path within the braces
    data1 = f1.readlines() #creates a list
    data1.pop(0) #removes header
    converted_list1 = []
    
    for element in data1:
        converted_list1.append(element.strip())
        
    data1 = "".join(converted_list1) 

with open('C:/Users/white/Desktop/seq2.fasta', 'r') as f2: #put your path within the braces
    data2 = f2.readlines() #creates a list
    data2.pop(0) #removes header
    converted_list2 = []
    
    for element in data2:
        converted_list2.append(element.strip())
        
    data2 = "".join(converted_list2)
    
    print("Length of data 1: "+ str(len(data1)))
    print("Length of data 2: "+ str(len(data2)))
    
# Input:  Two strings p and q
# Output: An integer value representing the Hamming Distance between p and q.
def HammingDistance(p, q):
    # your code here
    count = 0
    l = len(p)
    print("===============================================================")
    print("MUTATIONAL POSITION\t\tPATTERN 1\t\tPATTERN 2")
    print("===============================================================")
    for i in range (l):
        if p[i] != q[i]:
          print(" \t " + str(i) + "   \t\t\t " + p[i] + " \t\t\t "+ q[i])
          count += 1
    return count

p = data1
q = data2

print("Total Mismatches " + str(HammingDistance(p, q)))
```
